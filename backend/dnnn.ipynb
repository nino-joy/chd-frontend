```python
import numpy as np
import pandas as pd
import keras
```

```python
df=pd.read_csv('heart.csv')
df.head()
```

```python
X = df.drop(columns=['condition'])
Y = df[['condition']]
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = keras.models.Sequential()


model.add(Dense(12, activation='relu', input_shape=(13,)))
model.add(Dense(71, activation='relu'))
model.add(Dense(51, activation='relu'))
model.add(Dense(31, activation='relu'))
model.add(Dense(11, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(3, activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X, Y, epochs=1000, callbacks=[keras.callbacks.EarlyStopping(patience=3)])
```

```python
test_data = np.array([69,	1,	0,	160,	234,	1,	2,	131,	0,	0.1	,1,	1,	0])
print(model.predict(test_data.reshape(1,13), batch_size=1))
```
